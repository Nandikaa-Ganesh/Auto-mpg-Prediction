```python
sns.pairplot(df)
plt.show()
```

```python
# Scatterplot: A scatterplot shows the relationship between two variables
sns.scatterplot(x='origin', y='mpg', data=df)
plt.show()

sns.scatterplot(x='acceleration', y='mpg', data=df)
plt.show()

sns.scatterplot(x='model_year', y='mpg', data=df)
plt.show()

sns.scatterplot(x='cylinders', y='weight', data=df)
plt.show()
```

```python
# Boxplot: A boxplot shows the distribution of a continuous variable
sns.boxplot(x='acceleration', y='mpg', data=df)
plt.show()

sns.boxplot(x='origin', y='mpg', data=df)
plt.show()

sns.boxplot(x='model_year', y='mpg', data=df)
plt.show()
```

```python
# Violinplot: A violinplot shows the distribution of a continuous variable
sns.violinplot(x='model_year', y='mpg', data=df)
plt.show()
```

```python
# Histogram: A histogram shows the distribution of a continuous variable
sns.histplot(y='mpg', data=df)
plt.show()
```

```python
# Barplot: A barplot shows the relationship between a categorical and a continuous variable
sns.barplot(x='model_year', y='mpg', data=df)
plt.show()

sns.barplot(x='cylinders', y='mpg', data=df)
plt.show()

sns.barplot(x='origin', y='mpg', data=df)
plt.show()
```

```python
# Countplot: A countplot shows the count of each category in a categorical variable
sns.countplot(x='origin', data=df)
plt.show()
```

```python
# Heatmap: A heatmap shows the correlation between variables
sns.heatmap(df.corr(), annot=True, cmap='inferno', fmt='.2f')
plt.show()
```
